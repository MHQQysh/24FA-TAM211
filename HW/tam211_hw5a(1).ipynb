```python
# 4
import sympy as s
import matplotlib.pyplot as plt
import numpy as np
x=s.S('x')

w=3
l=3



fun1=x
f=s.integrate(fun1,(x,0,l))

funv=f-s.integrate(fun1,(x,0,x))
print(funv)


funv_lambdified = s.lambdify(x, funv, 'numpy')

# 生成绘图数据
x_vals = np.linspace(0, l, 400)  # 根据需要调整范围
y_vals = funv_lambdified(x_vals)

# 绘制图形
plt.plot(x_vals, y_vals, label='funv(x)')
plt.xlabel('x')
plt.ylabel('funv(x)')
plt.title('Plot of funv(x)')
plt.legend()
plt.grid(True)
plt.show()


funm=s.integrate(funv,(x,0,x))
print(funm)
funm = s.lambdify(x, funm, 'numpy')
x_vals = np.linspace(0, l, 10)  # 根据需要调整范围
y_vals = funm(x_vals)

plt.plot(x_vals, y_vals, label='funv(x)')
```
